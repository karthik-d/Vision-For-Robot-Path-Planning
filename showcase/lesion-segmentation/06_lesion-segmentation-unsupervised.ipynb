```python
import nibabel as nib
from matplotlib import pyplot as plt
from skimage.util import montage as montage2d

scan_img = nib.load('./data/Brats18_TCIA02_151_1/Brats18_TCIA02_151_1_t1ce.nii.gz').get_data()
fig, ax1 = plt.subplots(1, 1, figsize = (20, 20))
ax1.imshow(montage2d(scan_img), cmap ='bone')

tumor_img = nib.load('whole_tumor.nii.gz').get_data()
fig, ax2 = plt.subplots(1, 1, figsize = (20, 20))
ax2.imshow(montage2d(tumor_img), cmap ='bone')
# fig.savefig('ct_scan.png')
```

```python
print(scan_img.shape)

for idx in range(75, 111, 5):
    plt.figure()
    fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12, 6))
    ax1.imshow(scan_img[:,:,idx])
    ax1.set_title('Image')

    ax2.imshow(tumor_img[:,:,idx])
    ax2.set_title('Mask')
plt.show()
```
